```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
# Loading the dataset
df = pd.read_csv("C:\\Users\\User\\Desktop\\Drive D\\New folder\\ML\\Completed\\AirBnb.csv")
df.head()
```

```python
df.set_index("Date",inplace=True)
df
```

```python
x = df.drop(columns=['Adj Close','Volume'])
x
```

```python
y = df['Adj Close']
y
```

```python
#Splitting the dataset into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sgd_reg = SGDRegressor(random_state=42)
```

```python
sgd_reg.fit(x_train,y_train)
```

```python
pred = sgd_reg.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
sgd_param_grid = {
    'penalty' : ['l2','l1','elasticnet'],
    'alpha' : [0.00005,0.0001,0.00015],
    'l1_ratio' : [0.1,0.15,0.2],
    'max_iter' : [500,1000,1250],
    'learning_rate' : ['constant','optimal','invscaling','adaptive'],
    'eta0' : [0.005,0.01,0.02],
    'power_t' : [0.1,0.25,0.3],
}
```

```python
# Step 3: Create GridSearchCV object
grid_search = GridSearchCV(estimator=sgd_reg, param_grid=sgd_param_grid, cv=5, verbose=3)
```

```python
# Step 4: Fit the model
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
sgd_reg = SGDRegressor(alpha=0.0001,eta0=0.005,l1_ratio=0.1,learning_rate='adaptive',max_iter=500,penalty='l1',power_t=0.1,random_state=42)
sgd_reg.fit(x_train,y_train)
pred = sgd_reg.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```
