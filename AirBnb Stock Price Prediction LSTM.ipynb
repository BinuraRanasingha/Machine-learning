# Neural Network made using PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
```

```python
df
```

```python
# Normalize data
#scaler = MinMaxScaler(feature_range=(-1, 1))
vals = df['Adj Close'].values.reshape(-1,1)
#data_normalized = scaler.fit_transform(vals)
data_normalized = vals
```

```python
# Convert data to PyTorch tensor
data_tensor = torch.Tensor(data_normalized).unsqueeze(2)  # Add a third dimension (features)
```

```python
# Split data into training and testing sets
train_data = data_tensor[:622]
test_data = data_tensor[622:]
```

```python
# Define LSTM model
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, 
                            hidden_size=hidden_size, 
                            num_layers=num_layers, 
                            batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out
```

```python
# Initialize model
input_size = 1
hidden_size = 64
num_layers = 2
output_size = 1
model = LSTMModel(input_size, hidden_size, num_layers, output_size)
```

```python
# Loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop
num_epochs = 6775
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(train_data)
    
    # Reshape target data to match the shape of outputs
    target = train_data[:, -1, :].unsqueeze(1)
    target = target.reshape(-1,1)

    loss = criterion(outputs, target)

    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 1 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Set the model to evaluation mode
```

```python
# Set the model to evaluation mode
model.eval()

# Make predictions on the test data
with torch.no_grad():
    predictions = model(test_data)
    predictions_squeezed = torch.squeeze(predictions, dim=1)  # Squeeze the extra dimension

# Reshape target test data to match the shape of predictions
target_test = test_data.squeeze(2)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(target_test.numpy(), predictions_squeezed.numpy())

# Print MSE
print(f"Mean Squared Error on Test Data: {mse}")
```
