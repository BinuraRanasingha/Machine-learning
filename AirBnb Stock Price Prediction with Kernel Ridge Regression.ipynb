```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.kernel_ridge import KernelRidge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#Importing the dataset
df = pd.read_csv("C:\\Users\\User\\Desktop\\Drive D\\New folder\\ML\\Completed\\AirBnb.csv")
df.head(10)
```

```python
df.set_index("Date",inplace=True)
df
```

```python
x = df.drop(columns=["Adj Close","Volume"])
x
```

```python
y = df['Adj Close']
y
```

```python
#Splitting the dataset into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
kernel_ridge_model = KernelRidge()
```

```python
kernel_ridge_model.fit(x_train,y_train)
```

```python
pred = kernel_ridge_model.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test, pred)
```
